Loading and Cleaning Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

raw_data = pd.read_csv('C:/Users/barkha arora/Desktop/machine learning/countries/Countries-exercise.csv')

print(raw_data.head())
data = raw_data.copy()
print(raw_data.isnull().sum())  
raw_data = raw_data.dropna()    
data = raw_data.copy()

```

```python
plt.scatter(data['Longitude'], data['Latitude'])
plt.xlim(-180, 180)
plt.ylim(-90, 90)
plt.title('Geographical Location of Countries')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
x = data[['Longitude', 'Latitude']]
```

Data Normalization

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

Applying DBSCAN

```python
dbscan = DBSCAN(eps=0.3, min_samples=5)
dbscan.fit(x_scaled)


```

Understanding dbscan.labels

```python
labels = dbscan.labels_
```

```python
data_with_clusters = data.copy()
data_with_clusters['Cluster'] = dbscan.labels_
```

Visualizing Clusters

```python
plt.scatter(data['Longitude'], data['Latitude'], c=data_with_clusters['Cluster'], cmap='rainbow')
plt.show()

```
